```python
#bar graphs for categorical variables /gdp of country/
import matplotlib.pyplot as plt

x=[1,4,7,15]
y=[11,23,20,12]
x2=[2,3,9,13]
y2=[11,20,10,24]
plt.bar(x,y,label='Ex1')

plt.bar(x2,y2,label='Ex2')

plt.legend()
plt.title('Bar Plot')
plt.xlabel('Bar Number')
plt.ylabel('Bar Hieght')

plt.show()
```

```python
#histogram quatitative variables age gruop gdp of the population
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


population_age=np.arange(np.random.randint(0,100))
bins= np.arange(0.0,100,10)



plt.hist(population_age,bins,histtype='bar',rwidth=0.8)




plt.title('Hitogram Plot')
plt.xlabel('age')
plt.ylabel('population')
plt.legend()
plt.show()
```

```python
#to find the relationship between variable called finding correlation
import matplotlib.pyplot as plt

x=[1,4,7,5,3,8]
y=[2,3,5,6,7,9]

plt.scatter(x,y,label='Skitscat',color='k')



plt.legend()
plt.title('Scatter plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
#Area plot for tracking changes mutliple group/ variable which makes up into one hwole categories
import matplotlib.pyplot as plt

days=[1,2,3,4,5]
eating=[2,3,5,6,7]
sleeping =[2,4,5,6,7]

plt.plot([],[],color='b',label='eating',linewidth=5)
plt.plot([],[],color='g',label='sleeping',linewidth=5)

plt.stackplot(days,eating,sleeping,colors=['b','g'])



plt.legend()
plt.title('Area plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```
